```python
import numpy as np
import pandas as pd
import pickle
```

```python
df = pd.read_csv("./train.csv")
df.head()
```

```python
df = pd.concat([df.drop('BusinessTravel', axis=1), pd.get_dummies(df['BusinessTravel'],prefix="BT")], axis=1)
```

```python
df = pd.concat([df.drop('Department', axis=1), pd.get_dummies(df['Department'],prefix="Dep")], axis=1)
```

```python
df = pd.concat([df.drop('Gender', axis=1), pd.get_dummies(df['Gender'])], axis=1)
```

```python
df = pd.concat([df.drop('MaritalStatus', axis=1), pd.get_dummies(df['MaritalStatus'])], axis=1)
```

```python
df = pd.concat([df.drop('EducationField', axis=1), pd.get_dummies(df['EducationField'],prefix="Education")], axis=1)
```

```python
df.head()
```

```python
df = pd.concat([df.drop('OverTime', axis=1), pd.get_dummies(df['OverTime'],prefix="OT")], axis=1)
```

```python
df = pd.concat([df.drop('JobRole', axis=1), pd.get_dummies(df['JobRole'])], axis=1)
```

```python
df.info()
```

```python
df.drop(columns=["Id","EmployeeNumber"],inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#features = df[['Age','BT_Non-Travel','BT_Travel_Frequently','BT_Travel_Rarely','Healthcare Representative','Human Resources','Laboratory Technician','DistanceFromHome','Education','EducationField','EnvironmentSatisfaction','Gender','JobInvolvement','JobRole','JobSatisfaction','MaritalStatus','MonthlyIncome','NumCompaniesWorked','OverTime','PercentSalaryHike','PerformanceRating','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager','CommunicationSkill','Behaviour','Attrition',]]
```

```python
X=df.loc[:, df.columns != 'Attrition']
x=X.to_numpy()
x.shape
X.info()
```

```python
Y=df['Attrition']
y=Y.to_numpy()
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=55)
y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred_gnb = gnb.fit(X_train, y_train).predict(X_test)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred_gnb).sum()))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_gnb))
```

```python
from sklearn import tree
dec_clf = tree.DecisionTreeClassifier()
y_pred_dec = dec_clf.fit(X_train, y_train)
y_pred_dec_clf=y_pred_dec.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_dec_clf))
```

```python
from sklearn.metrics import roc_auc_score,accuracy_score
roc_auc_score(y_test,y_pred_dec_clf)
```
